```python
from pathlib import Path
import sys
import os

current_directory = Path(os.getcwd())
print(current_directory)
root_directory = current_directory.parent.parent
print(root_directory)

sys.path.append(str(root_directory))
```

```python
import json
import typing as t
import re
```

```python
results_dirs = [
    root_directory / "evaluations/CoqGym100/ChainOfThought/test_preceding-lemmas-only-8064b986fc9a4e81b22c5863c2287268",
    root_directory / "evaluations/Wigderson100/ChainOfThought/wigderson_test_preceding-lemmas-only-63ef557c60434592a1640dabe6da6870",
    root_directory / "evaluations/BB5100/ChainOfThought/coq_bb5_test_preceding-lemmas-only-f343bff0daca42a8adf9c6b35a8618f7",
    root_directory / "evaluations/PnVRocqLib100/ChainOfThought/pnvrocqlib_test_preceding-lemmas-only-cac514b1ba744da29d09e98325202975"
]
sample_fileses = [
    list(results_dir.glob("*.samples.json"))
    for results_dir in results_dirs
]
sample_files = [
    sample_file
    for sample_files in sample_fileses
    for sample_file in sample_files
]
print(f"number of sample files: {len(sample_files)}")
```

```python
class Sample(t.TypedDict):
    code: str
    success: bool

sampleses: t.List[t.List[Sample]] = [json.loads(f.read_text()) for f in sample_files]
samples: t.List[str] = [
    sample["code"]
    for samples in sampleses
    for sample in samples
    if sample["code"].strip() != ""
]
print(f"number of samples: {len(samples)}")
for sample in samples[:5]:
    print("# sample")
    print(sample)
    print()
```

```python
BULLET_REGEX = r"\s*[-+*]+\s*"

for sample in samples[:5]:
    if re.search(BULLET_REGEX, sample):
        print("has bullet")
    else:
        print("no bullet")
    print(sample)
    print()
```

```python
num_samples = len(samples)
num_samples_with_bullets = sum(1 for sample in samples if re.search(BULLET_REGEX, sample))
print(f"number of samples: {num_samples}")
print(f"number of samples with bullets: {num_samples_with_bullets}")
print(f"percentage of samples with bullets: {num_samples_with_bullets / num_samples * 100}")
```
