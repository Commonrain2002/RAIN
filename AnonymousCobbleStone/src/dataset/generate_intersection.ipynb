```python
example = COQGYM_TEST_SAMPLED_DATASET[0]

def project_name(example: Example) -> str:
    return example.location.project_name.split("/")[-1]

def file_name_json(example: Example) -> str:
    # theories/PBTree.v -> theories/PBTree.json
    return example.location.file_name.replace(".v", ".json")

def lemma_name(example: Example) -> str:
    return example.location.lemma_name

def intersection_path(example: Example) -> str:
    return f"{project_name(example)}/{file_name_json(example)}/{lemma_name(example)}"
```

```python
examples_by_project = {}
for example in COQGYM_TEST_SAMPLED_DATASET:
    pn = project_name(example)
    if pn not in examples_by_project:
        examples_by_project[pn] = []
    examples_by_project[pn].append(example)

for pn, examples in examples_by_project.items():
    print(pn, len(examples))
```

```python
intersection_paths_by_project = {}
for example in COQGYM_TEST_SAMPLED_DATASET:
    pn = project_name(example)
    if pn not in intersection_paths_by_project:
        intersection_paths_by_project[pn] = []
    intersection_paths_by_project[pn].append(intersection_path(example))

for pn, intersection_paths in intersection_paths_by_project.items():
    print(pn, len(intersection_paths))
```

```python
# dump to intersection.json
with open("intersection.coqgym100.json", "w") as f:
    json.dump(intersection_paths_by_project, f)
```
